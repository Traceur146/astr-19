```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
x = np.arange(0,1, 0.01)

def exp(x):
    return np.exp(x)
```

```python
y = exp(x)

plt.plot(x,y)
plt.xlabel('Time [milliseconds]')
plt.ylabel('Awesomeness')
plt.savefig('Session7.pdf', bbox_inches = 'tight', dpi = 600)
plt.show()
```
